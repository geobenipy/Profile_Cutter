```python
# -*- coding: utf-8 -*-
"""
Created in November 2023

@author: Benedikt Haimerl

Description: This script takes the Survey World Coordinate output from Kingdom and makes single Profiles out of it.
"""

import pandas as pd

# Einlesen der .dat-Datei
file_path = 'AL605.dat'  
data = pd.read_csv(file_path, delim_whitespace=True, header=None)

# Benennung der Spalten
data.columns = ['Profil', 'CMP', 'X_UTM', 'Y_UTM']

# Gruppieren der Daten nach dem Namen und Erstellen separater .dat-Files
for name, group in data.groupby('Profil'):
    output_file = f"{name}.dat"
    group.to_csv(output_file, sep="\t", index=False, header=False)

print("Fertig.")
```
